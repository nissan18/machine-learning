```python
# let's try to model function y=x with domain [0,100] and noise = +/-5
import numpy as np
N = 10  # sample size
X = np.random.randint(0, 101, N)
noise = np.random.rand(N) * 10 - 5
Y = X + noise
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot(np.arange(0,100))
plt.plot(X, Y, "ro")
plt.axis([0, 100, 0, 110])
plt.show()
```
